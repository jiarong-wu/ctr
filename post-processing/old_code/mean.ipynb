```python
profs_tensem = []
for t in ([38.6]):
    ds = pd.read_csv('test_NWP/prof_%g' %t, sep='\t', names=['z','ux','uy','uz','uxuy','uxux','uyuy','uzuz','a'], header=1)
    profs_tensem.append(ds)
```

```python
plt.plot(profs_tensem[0].z, profs_tensem[0].a)

Ustar = 0.25**2
L0 = 2*np.pi
h_ = 1
total = Ustar**2*0.5
def forcing (y):
    return 4*total / (L0-h_)**4 * (L0-h_ - 2*(y-h_))**2 * (5*(L0-h_) - 4*(y-h_))
def a(y):
    half = (L0-h_)/2.+1
    return [1 if elem > half else 0 for elem in y]
plt.plot(profs_tensem[0].z, forcing(profs_tensem[0].z)*a(profs_tensem[0].z))
```
